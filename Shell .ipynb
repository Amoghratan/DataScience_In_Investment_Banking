# Shell 

Major Commands in Shell, a quick reference

##### Current Working Directory
pwd

##### Relative Path - Doesnt start with '/' 
ls data

##### Absolute paths start with '/'
- ls /desktop/folder/data

##### Going a level up
cd ..

##### Going to home
cd ~

##### Copying file
cp originial.txt copied.txt
cp data/summer.csv backup

##### Moving file
mv data/summer.csv data/winter.csv data2

##### Renaming file
mv summer.csv summer_rename.csv data2

##### Removing files
rm summer.csv 

mkdir new_dir

Temp 
cd /tmp
mkdir scratch

##### Viewing a file content
cat file_name.txt

##### Viewing a file in part
less file_name.csv file_name2.csv
spacebar to go to next page
:n to go to next file 
:q to quit 

##### Display first 10 lines 
head file_name.txt

head -3 file_name.txt

ls -R : To see everything under a directory
ls -R -F : 
 

##### Display all but first 6 lines of the file
tail -n +7 file_name.csv : 

##### Get help on a command
man head
cut -f 2-5,8 -d , values.csv : select columns <br>

history : will display all commands used  <br>
!head : will repeat latest command using head <br> 
!2 : repeat 2 command <br>

head, tail : select rows <br>
cut : select columns  <br>
grep : select lines according to what they contain <br>
-c: print a count of matching lines rather than the lines themselves <br>
-h: do not print the names of files when searching multiple files <br>
-i: ignore case (e.g., treat "Regression" and "regression" as matches) <br>
-l: print the names of files that contain matches, not the matches <br>
-n: print line numbers for matching lines <br>
-v: invert the match, i.e., only show lines that don't match <br>

##### Searching string in txt file
grep word file_name.csv <br>
grep -v -n file_name.csv : display rows with line numbers in file <br>
grep -c hello filename1.csv filname2.csv : count number of lines with hello <br>

##### Storing a command output as a file 
head -n 10 filename.csv > temp.csv <br>

##### Combining Commands
Select 2nd column from a comma delimited file and pipe the result to grep with inverted match to exclude line containing the word "Tooth" <br>
cut -d , -f 2 file.csv ! grep -v word <br>

##### Count rows with dates in 2017 July 
grep 2017-07 file.csv | wc -l <br>

##### Seasonal has 4 files - Summer.csv, Winter.csv, Spring.csv, autum.csv
Print first 3 lines from both Summer and Spring <br>
head -n 4 seasonal/s*.csv

##### Wildcards
\*  All <br>
? Single Character <br>
[...]  Will match any 1 character 201[78].txt will match 2017.txt and 2018.txt <br>
{*.txt, *.csv}  Will match all files with txt and csv <br>

##### Sorts lines of text
cut -d , -f 2 seasonal/winter.csv | grep -v Tooth | sort -r

##### Removing Duplicates 
cut -d , -f 2 seasonal/winter.csv | grep -v Tooth | sort -r | uniq -c

##### Storing file in front of command
\> result.txt head -n 3 file.csv

##### Shell storing data 
Shell store data in variables like other languages, there are some known as enviornment variables which are written in upper case <br>
set | grep HOME <br>
HOME = /home/repl <br>

##### Variable values
echo Kshitiz Sharma <br>
echo $OSTYPE <br>

##### Shell Variable or local variable in programming language
testing=seasonal/winter.csv  <br>
head -n 1 $testing <br>

##### For loops
for var in docx pdf png;do echo $var;done <br>
for var in dir/*;do echo $var;done <br>

##### Similarity of loops 
Statement 1: grep -h 2017-07 seasonal/*.csv <br>
Statement 2: for f in seasonal/*.csv;do grep 2017-07 $f;done
